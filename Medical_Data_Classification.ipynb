# 2. Determine the frequency of distinct values in each feature set 

```python
def get_freq_continous(data,n,interval):
    a=[]
    for i in range(0,n):
        a.append(i)

    freq=[]
    for i in range(0,n):
        freq.append(0)

    for i in data:
        if i>=0:
            freq[int(i/interval)]+=1

    sum=0
    
    l=[]
    
    for i in range(0,n):
        l.append(i*interval)
        
    for i in range(0,n):
        sum+=freq[i]
        freq[i]/=data.shape[0] 
        print(l[i],"-",l[i]+interval,"\t",round(freq[i]*100,2),"\b%")
    print("TOTAL",sum/data.shape[0]*100,"\b%")
    print()
    
n=max(data2["Age"])/10+1
print("Age")
get_freq_continous(data1["Age"],int(n),10)
data1["Age"].hist(bins=int(n)+1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def get_freq(data,name):
    n=max(data)-min(data)+1
    a=[]
    for i in range(0,n):
        a.append(i)

    freq=[]
    for i in range(0,n):
        freq.append(0)

    for i in data:
        freq[i]+=1

    sum=0
    
    l=[]
    
    for i in range(0,n):
        l.append(i)
        
    print(name)
    for i in range(0,n):
        sum+=freq[i]
        freq[i]/=data.shape[0] 
        print(l[i],"\t",round(freq[i]*100,2),"\b%")
    print("TOTAL",sum/data.shape[0]*100,"\b%")
    print()

def get_freq_encoded(data,encoder):
    
    n=max(data)-min(data)+1
    #print(n)

    a=[]
    for i in range(0,n):
        a.append(i)

    freq=[]
    for i in range(0,n):
        freq.append(0)

    for i in data:
        freq[i]+=1

    sum=0
    l=encoder.inverse_transform(a)

    for i in range(0,n):
        sum+=freq[i]
        freq[i]/=data.shape[0] 
        print(l[i],"\t",round(freq[i]*100,2),"\b%")

    print("TOTAL",sum/data.shape[0]*100,"\b%")
    print()

get_freq(data2["Scholarship"],"Scholarship")
get_freq(data2["Hypertension"],"Hypertension")
get_freq(data2["Diabetes"],"Diabetes")
get_freq(data2["Alcoholism"],"Alcoholism")
get_freq(data2["SMS_received"],"SMS_received")
get_freq(data2["Handicap"],"Handicap")

label_encoder = LabelEncoder()
# Fit and transform the column
encoded_Gender = label_encoder.fit_transform(data2["Gender"])

print("Gender")
get_freq_encoded(encoded_Gender,label_encoder)

# Fit and transform the column
encoded_No_show = label_encoder.fit_transform(data2["No-show"])

print("No-show")
get_freq_encoded(encoded_No_show,label_encoder)

plt.subplot(3,3,1)
sns.countplot(x='Scholarship',data=data2)
plt.subplot(3,3,2)
sns.countplot(x='Hypertension',data=data2)
plt.subplot(3,3,3)
sns.countplot(x='Diabetes',data=data2)
plt.subplot(3,3,4)
sns.countplot(x='Alcoholism',data=data2)
plt.subplot(3,3,5)
sns.countplot(x='SMS_received',data=data2)
plt.subplot(3,3,6)
sns.countplot(x='Handicap',data=data2)
plt.subplot(3,3,7)
sns.countplot(x='Gender',data=data2)
plt.subplot(3,3,8)
sns.countplot(x='No-show',data=data2)
plt.subplots_adjust(wspace = 0.8,hspace = 0.8)
```

# 5. The values within AwaitingTime are negative, transform them into positive values

# Here we process the date information. We import datetime to caculate the awaiting time(we keep the days only). For further processing of the date information(i.e., standardization or normalization), we transform the feature ScheduledDay and AppointmentDay into days by subtracting a reference date(i.e. the earliest day of all date information) from them.

```python
import datetime

AwaitingTime=[]

SD=data3["ScheduledDay"]
AD=data3["AppointmentDay"]

Day_min=datetime.datetime.strptime(min(min(SD),min(AD))[0:10], '%Y-%m-%d')

SD2=[]
AD2=[]

for i,j in zip(SD,AD):
    d1 = datetime.datetime.strptime(i[0:10], '%Y-%m-%d')
    d2 = datetime.datetime.strptime(j[0:10], '%Y-%m-%d')
    delta=d2-d1
    delta1=d1 - Day_min
    delta2=d2 - Day_min
    SD2.append(delta1.days)
    AD2.append(delta2.days)
    AwaitingTime.append(delta.days)
    
data3["AwaitingTime"]=AwaitingTime
data3["ScheduledDay_New"]=SD2
data3["AppointmentDay_New"]=AD2
```

# partitioning the dataset into a train dataset (70%) and test dataset (30%)

```python
#Splitting the dataset into training and testing variables
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20, stratify=y)

#keeping 70% as training data and 30% as testing data.
```

# Use the train dataset to build the Naïve Bayes and the test dataset to evaluate how well the model generalizes to future results. 


```python
from sklearn.naive_bayes import GaussianNB
 
#Calling the Class
naive_bayes = GaussianNB()
 
#Fitting the data to the classifier
naive_bayes.fit(X_train , y_train)

#Predict on training data
y_predicted1 = naive_bayes.predict(X_train)

#Predict on test data
y_predicted = naive_bayes.predict(X_test)
```

# 1. Write a Function to detect the model’s Accuracy by applying the trained model on a testing dataset to find the predicted labels of Status. Was there overfitting?

```python
#Import metrics class from sklearn
from sklearn import metrics

print("Accuracy for training set:",metrics.accuracy_score(y_predicted1 , y_train))
print("Accuracy for test set:",metrics.accuracy_score(y_predicted , y_test))
```

# There's no overfitting since the accuracy on test set is not significantly lower than that on training set.

# 2. Tune the model using GridSearchCV

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

param_grid_nb = {
    'var_smoothing': np.logspace(0,-2, num=100)
}
nbModel_grid = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid_nb, verbose=1, cv=10, n_jobs=-1)
```

```python
best_model= nbModel_grid.fit(X_train, y_train)
```

```python
#Predict on training data
y_predicted1 = best_model.predict(X_train)

#Predict on test data
y_predicted = best_model.predict(X_test)

print("Accuracy for training set:",metrics.accuracy_score(y_predicted1 , y_train))
print("Accuracy for test set:",metrics.accuracy_score(y_predicted , y_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
print("Classification Report is:\n",classification_report(y_test,y_predicted))
print("\n F1:\n",f1_score(y_test,y_predicted))

print("\n Accuracy score is:\n",accuracy_score(y_test,y_predicted))
print("\n Precision score is:\n",precision_score(y_test,y_predicted))
print("\n Recall/Sensitivity score is:\n",recall_score(y_test,y_predicted))
tn, fp, fn, tp = confusion_matrix(y_test,y_predicted).ravel()
print("\n Specificity score is:\n",tn / (tn+fp))
roc_naive_bayes=metrics.RocCurveDisplay.from_predictions(y_test,y_predicted)
```
