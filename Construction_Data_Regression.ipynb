# To deal with the imbalance problem, we can apply SMOTE algorithm to the dataset. However, SMOTE is designed for classification problem, so we need to first bucket the data, which means, we can map the values of CompressiveStrength into different strength levels. 

# The following steps of dealing with the imbalance problem in this specific case has been regarded as a valid method by the professor in class.

# Here we first determine the level of CompressiveStrength for each sample, ranging from level 0 to 39, and each level has a width of around 2 Mpa.

```python
level=[]
for i in data3["CompressiveStrength"]:
    if i==max(data3["CompressiveStrength"]):
        level.append(39)
        continue
    level.append(int((i-2.33)/2.00675))

data3["level"]=level
```

```python
data3
```

```python
data3.describe()
```

# Check if a level has fewer than 6 samples, as we will apply SMOTE algorithm to balance the data later.

```python
ct_idx=[]
count=[]

for i in range(0,40):
    if data3["level"][(data3.level==i)].count()<6:
        ct_idx.append(i)
        count.append(data3["level"][(data3.level==i)].count())
    
```

```python
ct_idx # levels having fewer than 6 samples
```

```python
count
```

# Clear those that belong to the levels having fewer than 6 samples. (We will recombine them after completing SMOTE algorithm)

```python
a=(data3.level!=0)

for i in ct_idx:
    a&=(data3.level!=i)

data3_2=data3[a].reset_index(drop=True)
```

```python
data3_2
```

# Apply SMOTE algorithm to balance the data. 

# (Noting that SMOTE is designed for classification problems rather than regression ones. 

# However, since we have already determined the discrete strength level of CompressiveStrength for each sample, we can then apply the SMOTE algorithm.)

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

X=data3_2.drop(["CompressiveStrength","level"], axis=1)
y=data3_2["level"]

sm = SMOTE(random_state=42, k_neighbors=5)
data3_3, y_res = sm.fit_resample(X, y)
```

```python
y_res.shape
```

```python
data3_2["CompressiveStrength"]
```

```python
data3_3
```

# Next, we need to fill out the value of CompressiveStrength for each new sample generated by SMOTE algorithm. Here we use the midpoint value of each level. 

# For example, if the level of CompressiveStrength of the sample is 0, then we fill out its CompressiveStrength with value 3.33 Mpa, as the range of CompressiveStrength for level 0 is from 2.33 to 4.33 Mpa. 

# This way, we can ensure that the differences between the ground truths and the filled values are equal or less than 1 Mpa.

```python
y_res_1=[]
for i in y_res[data3_2.shape[0]:y_res.shape[0]]: # the new generated samples are concated to the original dataset after running SMOTE algorithm
    y_res_1.append(3.333375+i*2.00675)

data3_3["CompressiveStrength"]=np.array(data3_2["CompressiveStrength"]).tolist()+y_res_1
```

```python
data3_3
```

```python
y_res
```

```python
data3_3["level"]=y_res
```

# Recombine the data with those cleared before.

```python
b=(data3.level==0)

for i in ct_idx:
    b|=(data3.level==i)

data3_5=data3[b].reset_index(drop=True) # filter out samples belonging to levels that have fewer than 6 samples using the indices that have already been obtained.
```

```python
from sklearn.utils import shuffle
# Recombine the data
data3_6=shuffle(pd.concat([data3_3,data3_5])).reset_index(drop=True)#.drop(["level"],axis=1)
```

# III. Model Development I (30 marks): 

# Ensemble Method: 

# â€¢ Split dataset into train (70%) and test (30%) and build predictive models to determine the strength of concrete using the following techniques: K-Nearest Neighbor Regressor, Random Forest, and XGBoost

```python
y=data5["CompressiveStrength_std_scaler"]
X=data5.drop(["CompressiveStrength_std_scaler","level"],axis=1)
```

# Split dataset into train (70%) and test (30%)

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20, stratify=data5["level"])
```

```python
from math import sqrt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

# K-Nearest Neighbor Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

KNreg = KNeighborsRegressor(n_neighbors=5)

KNreg.fit(X_train, y_train)

print("Test set R^2: {:.2f}".format(KNreg.score(X_test, y_test)))
```

```python
y_predict = KNreg.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_predict))
print("MSE:", mean_squared_error(y_test, y_predict))
print("rmse:", sqrt(mean_squared_error(y_test, y_predict)))
print("r2 score:", r2_score(y_test, y_predict))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

RFreg = RandomForestRegressor(n_estimators = 100, random_state = 0)

# fit the regressor with x and y data
RFreg.fit(X_train, y_train)

print("Test set R^2: {:.2f}".format(RFreg.score(X_test, y_test)))
```

```python
y_predict = RFreg.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_predict))
print("MSE:", mean_squared_error(y_test, y_predict))
print("rmse:", sqrt(mean_squared_error(y_test, y_predict)))
print("r2 score:", r2_score(y_test, y_predict))
```

```python
y_predict = XGBreg.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_predict))
print("MSE:", mean_squared_error(y_test, y_predict))
print("rmse:", sqrt(mean_squared_error(y_test, y_predict)))
print("r2 score:", r2_score(y_test, y_predict))
```
